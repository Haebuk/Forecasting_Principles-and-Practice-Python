```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
tute1 = pd.read_csv('../data/tute1.csv').drop(['Unnamed: 0'], axis=1)
tute1.head()
```

```python
date = pd.date_range('1981', periods=len(tute1), freq='Q')
tute1.set_index(date, inplace=True)
tute1.head()
```

```python
plt.subplots(3, 1, figsize=(10, 10))
for i in range(3):
    plt.subplot(3, 1, i+1)
    plt.plot(tute1.iloc[:, i])
    plt.title(tute1.columns[i])
plt.tight_layout()
```
